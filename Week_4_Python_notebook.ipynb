**Cleaning the data**

Validity

```python
# dropping data that is irrelevant from the dataset

df.drop(['Cars','Charge Slots','Charging Status','Displayed comment','Geo point','Rental status','Scheduled at','Slots','Station type','Subscription status'], axis = 1, inplace = True)
df
```

Completeness

```python
# check if there are null values on the dadatset

df.isnull().sum()
df
```

Consistency

```python
# checking for duplicate values 

df.duplicated().sum()
df
```

Uniformity

```python
# change column names to uppercase 
# removing decimal point on column name
# removing spaces in column names and replacing them with _

df.columns = df.columns.str.strip().str.replace('.','').str.upper().str.replace(' ','_')
df
```

**Finding Outliers**

```python
# create the BLUECAR_COUNTER's box plot

df.boxplot(column = ['BLUECAR_COUNTER'], grid = False)
```

```python
# creating the UTILIB_COUNTER's box plot

df.boxplot(column = ['UTILIB_COUNTER'], grid = False)
```

```python
# creating UTILIB_14_COUNTER's box plot

df.boxplot(column = ['UTILIB_14_COUNTER'], grid = False)
```

Deleting Outliers

```python
# calculate the first quartile
Q1 = df.quantile(0.25)

# calculate the third quartile
Q3 = df.quantile(0.75)

# calculate the interquartile range
IQR = Q3-Q1

# remove the outliers and store them in a new dataset
df_new=df[~((df < (Q1-1.5*IQR))|(df > (Q3+1.5*IQR))).any(axis = 1)]

#preview our new dataset
df_new
```

```python
# displaying the shape of the data

df_new.shape
```

**Analysing the data**

```python
# creating usage columns for BLUECAR, UTILIB_COUNTER AND UTILIB_1.4_COUNTER

pd.options.mode.chained_assignment = None
df_new['BLUECAR_USAGE'] = df.BLUECAR_COUNTER.diff()
df_new['UTILIB_USAGE'] = df.UTILIB_COUNTER.diff()
df_new['UTILIB_1.4_USAGE'] = df.UTILIB_14_COUNTER.diff()
df_new.head(10)
```

Challenges

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar)
# in the city of Paris over the month of April 2018. 

df_paris = df_new.loc[(df_new['CITY'] == 'Paris')&(df_new['BLUECAR_USAGE'] < 0.0)]
df_paris
df_hour = df_paris.groupby('HOUR')['HOUR'].count().sort_values(ascending = False)
df_hour.head(1)
```

```python
# What is the most popular hour for returning cars?

df_return = df_new.loc[(df_new['BLUECAR_USAGE'] > 0.0) & (df_new['CITY'] == 'Paris')]
df_return
df_hour = df_return.groupby('HOUR')['HOUR'].count().sort_values(ascending = False)
df_hour.head(1)
```
